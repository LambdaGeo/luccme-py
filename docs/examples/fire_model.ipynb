```python
from luccmepy import Component, Model, IPlotMap, create_regular_grid, fill_regular_grid
from luccmepy.models.toys import FireModelProb, FireModel
```

```python
from matplotlib.colors import ListedColormap
custom_cmap = ListedColormap(['green', 'red', 'brown'])
```

## Rodando exemplo 1

```python
FOREST = 0
BURNING = 1
BURNED = 2

env = Model(
    gdf = create_regular_grid (bounds=(10,10,50,50), dim= 20, attrs= {"state": FOREST}),
    endTime = 10
)

env.gdf.loc["10-10","state"] = BURNING


IPlotMap(
    plot_params={ "column": "state","cmap": custom_cmap,  "ec" : 'black'}
)


FireModel(create_neighbohood="Rook")

env.run(10)
```
